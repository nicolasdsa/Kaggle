## Objective

Inspect the processed set-level dataset and understand the distribution of `num_parts`.


```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

DATA_PATH = Path("../data/processed/set_level_features.csv")
df = pd.read_csv(DATA_PATH)
df.head()
```

```python
# TODO: extend this notebook with the rest of your EDA.
# Basic expected result for this objective:
# - understand the shape of the dataset
# - inspect summary statistics for num_parts
# - visualize whether num_parts is skewed, dispersed, or has outliers

print(df.shape)
print(df["num_parts"].describe())

sns.histplot(df["num_parts"], bins=30)
plt.title("Distribution of num_parts")
plt.xlabel("num_parts")
plt.ylabel("count")
plt.show()
```
